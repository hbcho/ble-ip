```python
import csv
import json


sensors = []
with open('AA locations 2.txt', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        device = row[0]        
        sensors.append(
            {
                'id': device,
                'x': int(row[1]),
                'y': int(row[2])
            }
        )
```

```python
import numpy as np

N = len(sensors)
X = np.zeros((N,2))
y = np.zeros(N)
labels = []
for i, sensor in enumerate(sensors):
    # print(i, sensor)
    X[i,0] = sensor['x']
    X[i,1] = sensor['y']
    labels.append(sensor['id'][-2:])
```
